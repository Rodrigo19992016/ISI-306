```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])

plt.scatter(X[:,0], X[:,1])
```

```python
pca = PCA(n_components=2)
pca.fit(X)
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
pca.explained_variance_
```

```python
pca.singular_values_
```
